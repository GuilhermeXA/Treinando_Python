```python
'''
1
a Questão (2,0 pontos)
Faça um programa que leia da entrada padrão nomes de três arquivos texto. Os dois primeiros 
arquivos são existentes, chamados aqui de “entrada1.txt” e “entrada2.txt”, contendo em cada 
linha a pontuação de um jogador e seu respectivo nome, ordenados em ordem decrescente 
de pontuação. O terceiro arquivo deve ser criado e deverá ficar ordenado em ordem também 
decrescente de pontuação, mantendo todas as pontuações e respectivos nomes dos 
jogadores dos dois primeiros arquivos. Ao final do programa mostre o conteúdo do arquivo 
produzido.
Restrição
Nenhum arquivo poderá ser mantido por inteiro na memória principal, seja em lista ou outra 
variável qualquer.
'''
# Função para criar um gerador que lê um arquivo linha a linha em ordem decrescente de pontuação
def le_arquivo_decrescente(arquivo):
    linhas = arquivo.readlines()
    linhas.sort(key=lambda linha: int(linha.split()[0]), reverse=True)
    for linha in linhas:
        yield linha

# Ler os nomes dos arquivos de entrada
nome_arquivo1 = input("Digite o nome do primeiro arquivo: ")
nome_arquivo2 = input("Digite o nome do segundo arquivo: ")

# Abrir os dois arquivos de entrada para leitura
with open(nome_arquivo1, 'r') as arquivo1, open(nome_arquivo2, 'r') as arquivo2:
    # Criar geradores para ler os arquivos em ordem decrescente de pontuação
    gerador1 = le_arquivo_decrescente(arquivo1)
    gerador2 = le_arquivo_decrescente(arquivo2)

    # Abrir o terceiro arquivo para escrita
    nome_arquivo3 = input("Digite o nome do terceiro arquivo: ")
    with open(nome_arquivo3, 'w') as arquivo3:
        for linha1, linha2 in zip(gerador1, gerador2):
            # Dividir a linha em pontuação e nome
            pontuacao1, nome1 = linha1.split()
            pontuacao2, nome2 = linha2.split()

            # Escrever as linhas no arquivo de saída
            arquivo3.write(f"{pontuacao1} {nome1}")
            arquivo3.write(f"{pontuacao2} {nome2}")

        # Escrever qualquer conteúdo restante do primeiro arquivo (caso um arquivo tenha mais linhas que o outro)
        for linha1 in gerador1:
            arquivo3.write(linha1)

        # Escrever qualquer conteúdo restante do segundo arquivo
        for linha2 in gerador2:
            arquivo3.write(linha2)

print(f"Conteúdo do arquivo {nome_arquivo3}:")
with open(nome_arquivo3, 'r') as arquivo3:
    for linha in arquivo3:
        print(linha, end='')
```

```python
'''
2
a Questão (4,0 pontos)
Faça um programa que leia um arquivo texto escolhido pelo usuário, com várias palavras por 
linha separadas por espaço em branco e escreva:
(a) O conjunto das palavras contidas no arquivo, ordenadas em ordem alfabética crescente.
(b) Qual palavra que mais ocorreu. Caso haja empate, escreva uma delas.
Restrição
Não podem ser utilizadas rotinas de ordenação prontas e disponíveis na API da linguagem 
Python. Você deve implementar um dos algoritmos de ordenação vistos em aula
'''
# Função de ordenação por seleção
def ordenar_palavras(lista_palavras):
    for i in range(len(lista_palavras)):
        min_idx = i
        for j in range(i + 1, len(lista_palavras)):
            if lista_palavras[j] < lista_palavras[min_idx]:
                min_idx = j
        lista_palavras[i], lista_palavras[min_idx] = lista_palavras[min_idx], lista_palavras[i]

# Ler o nome do arquivo escolhido pelo usuário
nome_arquivo = input("Digite o nome do arquivo: ")

# Abrir o arquivo para leitura
try:
    with open(nome_arquivo, 'r') as arquivo:
        conteudo = arquivo.read()

        # Dividir as palavras por espaço em branco
        palavras = conteudo.split()

        # Criar um conjunto para armazenar as palavras únicas
        conjunto_palavras = set(palavras)

        # Criar uma lista para contagem de ocorrências
        lista_ocorrencias = []

        for palavra in conjunto_palavras:
            lista_ocorrencias.append((palavra, palavras.count(palavra)))

        # Ordenar a lista de palavras em ordem alfabética
        ordenar_palavras(conjunto_palavras)

        # Encontrar a palavra que mais ocorreu
        palavra_mais_ocorrencias = max(lista_ocorrencias, key=lambda x: x[1])[0]

        # Exibir o conjunto ordenado e a palavra que mais ocorreu
        print("Conjunto de palavras ordenadas em ordem alfabética:")
        for palavra in conjunto_palavras:
            print(palavra)
        print("Palavra que mais ocorreu:", palavra_mais_ocorrencias)

except FileNotFoundError:
    print("Arquivo não encontrado.")
```

```python
'''
3
a Questão (4,0 pontos)
Escreva um programa que receba como entrada os arquivos binários “entrada1.bin” e 
“entrada2.bin”, contendo em cada um deles uma coleção de números inteiros ordenados 
de forma crescente. Seu programa deve ler o conteúdo desses arquivos e produzir como 
saída o arquivo binário “saida.bin” contendo a coleção completa de números lidos, também 
ordenada.
Observação
É garantido que os arquivos de entrada estão ordenados.
Restrição
Você deve assumir que os arquivos de entrada são tão grandes que é impossível manter todo 
o conteúdo na memória principal do computador.
Exemplo
O exemplo que segue mostra o conteúdo dos arquivos binários de forma textual apenas para 
facilitar a interpretação dos dados. Você deve estar ciente que os arquivos “entrada1.bin”, 
“entrada2.bin” e “saída.bin” são binários.
'''
# Abrir os arquivos de entrada e saída
with open("entrada1.bin", "rb") as arquivo1, open("entrada2.bin", "rb") as arquivo2, open("saida.bin", "wb") as arquivo_saida:
    # Ler os primeiros números dos arquivos de entrada
    num1 = int.from_bytes(arquivo1.read(4), byteorder='big', signed=True)
    num2 = int.from_bytes(arquivo2.read(4), byteorder='big', signed=True)

    # Loop de mesclagem
    while True:
        if num1 < num2:
            # Escrever num1 no arquivo de saída
            arquivo_saida.write(num1.to_bytes(4, byteorder='big', signed=True))
            # Tentar ler o próximo número do arquivo1
            prox_num1 = arquivo1.read(4)
            if not prox_num1:
                # Se arquivo1 acabou, escrever num2 e o restante do arquivo2 no arquivo de saída
                arquivo_saida.write(num2.to_bytes(4, byteorder='big', signed=True))
                arquivo_saida.write(arquivo2.read())
                break
            num1 = int.from_bytes(prox_num1, byteorder='big', signed=True)
        else:
            # Escrever num2 no arquivo de saída
            arquivo_saida.write(num2.to_bytes(4, byteorder='big', signed=True))
            # Tentar ler o próximo número do arquivo2
            prox_num2 = arquivo2.read(4)
            if not prox_num2:
                # Se arquivo2 acabou, escrever num1 e o restante do arquivo1 no arquivo de saída
                arquivo_saida.write(num1.to_bytes(4, byteorder='big', signed=True))
                arquivo_saida.write(arquivo1.read())
                break
            num2 = int.from_bytes(prox_num2, byteorder='big', signed=True)

print("Mesclagem concluída. O arquivo de saída 'saida.bin' contém os números ordenados.")
```
